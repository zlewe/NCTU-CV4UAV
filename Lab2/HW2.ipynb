```python
%matplotlib inline

import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def showrgb(img):
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

# Q1

```python
image = cv2.imread('mj.tif')
```

```python
showrgb(image)
```

```python
image.shape
```

```python
mj_grayscale = image[:,:,0]
```

```python
def get_histogram(img):
    histogram = [0] * 256
    
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            histogram[img[i][j]] += 1
            
    plt.ylim(top=500)
    plt.plot(histogram)
    
    return histogram
```

```python
histogram = [0] * 256
```

```python
for i in range (mj_grayscale.shape[0]):
    for j in range (mj_grayscale.shape[1]):
        histogram[mj_grayscale[i][j]] += 1
```

```python
plt.plot(histogram)
```

```python
acc = [0] * 256
for i in range (256):
    if i == 0:
        acc[i] = (histogram[i]/563000)
    else:
        acc[i] = (histogram[i]/563000)+acc[i-1]
    
quan = [0] * 256
for i in range (256):
    quan[i] = int(acc[i]*255)
    
```

```python
mj_new = np.zeros(mj_grayscale.shape[:2],np.uint8)
for i in range(mj_grayscale.shape[0]):
    for j in range(mj_grayscale.shape[1]):
        mj_new[i][j] = quan[mj_grayscale[i][j]]
```

```python
showrgb(mj_new)
```

# Q2

```python
mask1 = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
mask2 = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
```

```python
def sobel_convolution(img,mask,threshold):
    rows = img.shape[0]
    column = img.shape[1]
    
    img_new = np.zeros([rows-2,column-2],np.uint8)
    
    for i in range(rows-2):
        for j in range(column-2):
            value = np.sum(np.multiply(mask, img[i:i+3,j:j+3]))
            if value < 0:
                img_new[i][j] = 0
            else:
                img_new[i][j] = value;
    
    histogram = get_histogram(img_new)
    
    for i in range(img_new.shape[0]):
        for j in range(img_new.shape[1]):
            if img_new[i][j] < threshold:
                img_new[i][j] = 0
            else:
                img_new[i][j] = 255
    
    return img_new
```

```python
hb = cv2.imread('HyunBin.jpg')
hb = hb[:,:,0]
```

```python
showrgb(hb)
```

```python
gx = sobel_convolution(hb,mask1,80)
```

```python
gy = sobel_convolution(hb,mask2,80)
```

```python
showrgb(gx)
```

```python
showrgb(gy)
```

```python
hb_edge = np.zeros(gx.shape[:2], np.uint8)
for i in range(gx.shape[0]):
    for j in range(gx.shape[1]):
        if gx[i][j] == 255 or gy[i][j] == 255:
            hb_edge = 255
```
